Developing a Model

```python
maxTc=max(Y)
targets=Y/1
```

 splitting data into train and test

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,targets,test_size=0.2, random_state=40)
```

```python
print(Xtrain)
print(Xtest)
```

```python
print(Ytrain)
print(Ytest)
```

Writing function to visualize our model

```python
from sklearn.metrics import mean_squared_error as mse
def performanceCalculator(absoluteValue,predictedValue,name):
    plt.plot([0,0.001,0.01,1],[0,0.001,0.01,1], color='blue')
    plt.scatter(absoluteValue,predictedValue, color='red')
    er=mse(absoluteValue,predictedValue)
    er=pow(er,0.5)
    er=int(er*10000)/10000
    plt.title('RMSE : ' +str(er) + 'for' + name)
    plt.show()
```

```python
# Lets start our Analysis with Linear Regression
from sklearn.linear_model import LinearRegression
linear_Regressor=LinearRegression()
linear_Regressor.fit(Xtrain,Ytrain)
```

```python
linear_Predictions=linear_Regressor.predict(Xtest)
performanceCalculator(Ytest,linear_Predictions,'Linear Regression')
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
r2l = r2_score(Ytest,linear_Predictions)
print(r2l)
```

```python
rmse_linear = np.sqrt(mean_squared_error(Ytest,linear_Predictions))
print(rmse_linear)
```

```python
# By using Decision Tree
from sklearn.tree import DecisionTreeRegressor
decision_Regressor=DecisionTreeRegressor()
decision_Regressor.fit(Xtrain,Ytrain)
```

```python
decision_predictions=decision_Regressor.predict(Xtest)
print(decision_predictions)
```

```python
performanceCalculator(Ytest,decision_predictions,'Decission Tree Regressor')
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
r2d = r2_score(Ytest,decision_predictions)
print(r2d)
```

```python
rmse_decision = np.sqrt(mean_squared_error(Ytest,decision_predictions))
print(rmse_decision)
```

```python
# By using Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
random_regressor=RandomForestRegressor()
random_regressor.fit(Xtrain,Ytrain)
```

```python
random_predictions=random_regressor.predict(Xtest)
print(random_predictions)
```

```python
performanceCalculator(Ytest,random_predictions,'Random Forest Regressor')
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
r2ra = r2_score(Ytest,random_predictions)
print(r2ra)
```

```python
rmse_random = np.sqrt(mean_squared_error(Ytest,random_predictions))
print(rmse_random)
```

```python
# By using Bagging classifier
from sklearn.ensemble import BaggingRegressor
bagging_regressor=BaggingRegressor()
bagging_regressor.fit(Xtrain,Ytrain)
```

```python
bagging_predictions=bagging_regressor.predict(Xtest)
print(bagging_predictions)
```

```python
performanceCalculator(Ytest,bagging_predictions,'Bagging Regressor')
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
r2b = r2_score(Ytest,bagging_predictions)
print(r2b)
```

```python
rmse_bagging = np.sqrt(mean_squared_error(Ytest,bagging_predictions))
print(rmse_bagging)
```

```python
# By using ensembling
prediction1=random_regressor.predict(Xtest)
prediction2=bagging_regressor.predict(Xtest)
predictions=(prediction1+prediction2)/2
```

```python
performanceCalculator(Ytest,predictions,'(Random Forest + Bagging)')
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
r2ra_b = r2_score(Ytest,predictions)
print(r2ra_b)
```

```python
rmse_Random_Bagging = np.sqrt(mean_squared_error(Ytest,bagging_predictions))
print(rmse_Random_Bagging)
```

From these models, Random Forest is the best model

```python
Y_lab=Ytest
predValue=random_regressor.predict(Xtest)
diff=abs(Y_lab - predValue)
ans_data={'True Tc' : Y_lab,
         'Predicted Tc ': predValue,
         'Absolute error ': diff}
ans_data=pd.DataFrame(ans_data)
ans_data
```
